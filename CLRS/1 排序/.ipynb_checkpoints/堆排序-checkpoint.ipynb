```python
# heapify
#  
def MAX_HEAPIFY(A, heap_size, i):
    # i
    left  = 2 * i + 1
    right = 2 * i + 2
    # 
    max = i
    if left  < heap_size and A[left]  > A[i]  : max = left  # 
    if right < heap_size and A[right] > A[max]: max = right
    # i
    if max != i:
        temp = A[i]
        A[i] = A[max]
        A[max] = temp
        MAX_HEAPIFY(A, heap_size, max)
```

```python
A = [9, 14, 10, 8, 7, 9, 3, 2, 4, 1]
MAX_HEAPIFY(A, len(A), 0)
A
```

```python
A = [16, 3, 10, 8, 7, 9, 3, 2, 4, 1]
MAX_HEAPIFY(A, len(A), 1)
A
```

```python
# 
def BUILD_MAX_HEAP(A):
    # MAX_HEAPIFY
    first_parent = int(len(A)/2) - 1
    for i in range(first_parent, -1, -1):
        MAX_HEAPIFY(A, len(A), i)
```

```python
A = [16, 3, 10, 8, 1, 9, 14, 2, 4, 7]
BUILD_MAX_HEAP(A)
A
```

```python
# 
# HEAPIFY
def HEAP_SORT(A):
    # 
    BUILD_MAX_HEAP(A)
    
    # HEAPIFY
    heap_size = len(A)-1
    while heap_size > 0:
        temp = A[0]
        A[0] = A[heap_size]
        A[heap_size] = temp
        
        MAX_HEAPIFY(A, heap_size, 0)
        
        heap_size -= 1
```

```python
A = [16, 3, 10, 8, 1, 9, 14, 2, 4, 7]
HEAP_SORT(A)
A
```
