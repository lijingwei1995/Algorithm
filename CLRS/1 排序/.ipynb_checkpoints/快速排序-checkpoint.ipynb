```python
def QUICK_SORT(A, L, R):
    # 
    # 
    # 
    # 1
    if L < R:
    
        # 
        # 
        x = A[R]
        
        i = L
        while i <= R-1: # 
            if A[i] > x: #  >= 
                #  j
                j = i + 1
                while j <= R-1:
                    if A[j] <= x:
                        # 
                        temp = A[i]
                        A[i] = A[j]
                        A[j] = temp
                        break
                    else:
                        j += 1
                # j break
                if j == R:
                    break
            else:
                i += 1
        # A[i]
        temp = A[i]
        A[i] = A[R]
        A[R] = temp
        
        # 
        QUICK_SORT(A, L, i-1)
        QUICK_SORT(A, i+1, R)
        
    return A
```

```python
A = [2,5,3,8,5,1,0]
QUICK_SORT(A, 0, len(A)-1)
```

```python
def reOrderArray(array):
    L = len(array)
    # i
    i = 0
    while i < L:
        if array[i] % 2 == 0:
            # j
            j = i + 1
            while j < L:
                if array[j] % 2 == 1:
                    # 
                    temp = array[j]
                    for k in range(j, i, -1):
                        array[k] = array[k-1]
                    array[i] = temp
                    break
                else:
                    j += 1
            if j == L:
                break
        else:
            i += 1
            
    return array
```

```python
reOrderArray([5, 4, 6, 8, 9, 2])
```
