题目描述
输入一个复杂链表（每个节点中有节点值，以及两个指针，一个指向下一个节点，另一个特殊指针指向任意一个节点），返回结果为复制后复杂链表的head。（注意，输出结果中请不要返回参数中的节点引用，否则判题程序会直接返回空）


```python
def Clone(pHead):
    if not pHead:
        return None
    
    # 复制头结点
    pHead_ = RandomListNode(pHead.label)
    
    p, p_ = pHead, pHead_
    q = p.next
    
    d = {p : p_} #建立两个链表之间的映射
    
    # 复制主链
    while q:
        q_ = RandomListNode(q.label)
        p_.next = q_
        d[q] = q_
        p = q
        q = q.next
        p_ = q_
        
    # 循环添加支链
    p, p_ = pHead, pHead_
    while p:
        p_.random = d[p.random]
        p = p.next
        p_ = p_.next
    
    return pHead_
        
        
```

```python
class RandomListNode:
    def __init__(self, x):
        self.label = x
        self.next = None
        self.random = None

a = RandomListNode(1)
b = RandomListNode(2)

d = {"1":a, b:"2"}
d["1"], d[b]
```
