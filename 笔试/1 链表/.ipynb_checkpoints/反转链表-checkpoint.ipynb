# 反转链表

题目描述
输入一个链表，反转链表后，输出新链表的表头



```python
import LinkedList
```

```python
def printList(listNode):
    r = []
    while listNode:
        r.append(listNode.val)
        listNode = listNode.next
    
    return r
```

```python
def ReverseList(pHead):
    
    if not pHead:
        return None
    
#     if not pHead.next:
#         return pHead
    
    p = pHead
    q = p.next
    p.next = None
    
    while q:
        r = q.next
        q.next = p
        p = q
        q = r
    
    return p
        
```

```python
l = LinkedList.LinkedList([4, 1, 2, 3, 8])
h = ReverseList(l.root)
printList(h)
```

经验：反转链表时，不是每次都将p的next置为None，因为第二次p变成了q，再删除会把原来的信息删除，所以只需第一次置为None即可。
